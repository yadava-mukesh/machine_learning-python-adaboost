```python
inp=df.drop('class', axis=1)
out=df['class']
```

```python
X=pd.get_dummies(inp, drop_first=True)
y=pd.get_dummies(out, drop_first=True)
```

***Train/Test Split***

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

# Adaboost

***HPT for best parameters***

```python
from sklearn.model_selection import GridSearchCV

# model/estimator
from sklearn.ensemble import AdaBoostClassifier
estimator_ab=AdaBoostClassifier()

#parameters which you want to tune 
param_grid_ab={'n_estimators':list(range(1,51))}

# GridSearchCV
grid=GridSearchCV(estimator_ab, param_grid_ab, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_estimator_.feature_importances_
```

```python
s1=pd.DataFrame(grid.best_estimator_.feature_importances_, columns=['feature'], index=X_train.columns.tolist())
s1
```

```python
# filter out important features
important_features = s1[s1['feature']>0].index.tolist()
important_features
```

```python
X_imp=X[important_features]
y=df['class']

from sklearn.model_selection import train_test_split
X_imp_train, X_imp_test, y_train, y_test = train_test_split(X_imp, y, test_size=0.2, random_state=101)

from sklearn.ensemble import AdaBoostClassifier
ab_best_model = AdaBoostClassifier(n_estimators=20) 
ab_best_model.fit(X_imp_train, y_train)

# train accuracy
ypred_train=ab_best_model.predict(X_imp_train)
from sklearn.metrics import accuracy_score
print('Train accuracy : ', accuracy_score(y_train, ypred_train))

# cross validation score
from sklearn.model_selection import cross_val_score
print('cross validation score : ', cross_val_score(ab_best_model, X_imp_train, y_train, cv=5).mean())

# test predict
ypred_test=ab_best_model.predict(X_imp_test)
from sklearn.metrics import accuracy_score
print('Test accuracy : ', accuracy_score(y_test, ypred_test))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, ypred_test)
```

```python
# Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, ypred_test))
```
